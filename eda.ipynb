```python
def balanced_review_sampling(df, book_id, max_reviews=50):
    book_reviews = df[df['book_id'] == book_id]
    sampled = []

    # group by rating
    for rating, group in book_reviews.groupby('rating'):
        # proportional share
        n = max(1, int(len(group) / len(book_reviews) * max_reviews))
        # take top-n sorted by votes (if column exists)
        sampled.extend(group.sort_values("n_votes", ascending=False).head(n).to_dict(orient="records"))
    
    return sampled

sampled_reviews = balanced_review_sampling(df_reviews, book_id="6936382", max_reviews=50)
```

```python
combined_reviews_romance = pd.DataFrame()

for id in book_ids :

    sampled_reviews = balanced_review_sampling(df_reviews, book_id=str(id), max_reviews=50)

    sampled_reviews = pd.DataFrame(sampled_reviews)

    combined_reviews_romance = pd.concat([combined_reviews_romance, sampled_reviews], axis = 0)

    
print (combined_reviews_romance.shape)
```
